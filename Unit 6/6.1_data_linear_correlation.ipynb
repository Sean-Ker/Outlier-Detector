```python
import matplotlib.pyplot as plt
import pylab
import pandas as pd
import numpy as np
plt.style.use('ggplot')
```

```python
# 3.
df = pd.DataFrame(zip([10,11,15,14,8,5],
                      [8,7,4,3,9,10],
                      [72,67,81,93,54,66]), 
                      columns=['Hours Studied','Hours Watching TV','Examination Score'])
df                
```

```python
# a.
# calc the trendline
x = df[['Hours Studied']]
m, b = np.polyfit(x.values.reshape(-1,), df.iloc[:,2], 1)
# Plot
df.plot.scatter(x='Hours Studied',y='Examination Score',title = 'Hours Studied vs. Examination Score', figsize=(10,6))
plt.plot(x,b+m*x,'-')
# plt.ylim(top=100)

# the line equation:
# print("y = %.6fx + %.6f"%(z[0],z[1]))
```

```python
print('Correlation Matrix:')
df.iloc[:,[0,2]].corr()
```

```python
# b.
# calc the trendline
x = df[['Hours Watching TV']]
m, b = np.polyfit(x.values.reshape(-1,), df.iloc[:,2], 1)
# Plot
df.plot.scatter(x='Hours Watching TV',y='Examination Score',title = 'Examination Score vs. Hours Studied', figsize=(10,6))
plt.plot(x,b+m*x,'-')
# plt.ylim(top=100)
# the line equation:
# print("y = %.6fx + %.6f"%(z[0],z[1]))
print('Correlation Matrix:')
df.iloc[:,[1,2]].corr()
```
